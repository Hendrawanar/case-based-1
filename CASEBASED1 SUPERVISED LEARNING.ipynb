```python
#membaca data dan import library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import math
import seaborn as sns

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/arrhythmia/arrhythmia.data'
df = pd.read_csv(url, header=None)
df.head()
```

```python
#mengurangi atribut
kolom = 15
while(kolom <= 278):
  df.drop(kolom, axis='columns',inplace = True)
  kolom += 1
df.head()
```

```python
#memberi nama kolom
df.columns = ['age', 'sex', 'height', 'weight', 'QRS duration', 'P-R Interval', 'Q-T Interval', 'T Interval', 'P interval', 'QRS', 'T', 'P', 'QRST', 'J', 'Heart rate', 'y']
df.head()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

mask = np.triu(np.ones_like(df.corr(), dtype=bool))

heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True)

plt.suptitle("Correlation Plot", ha='left', x=0.35, y=1.04, fontsize=20, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
#mengganti data yang berisi '?' dengan median
dfmean = df[(df['J'] != '?') & (df['QRST'] != '?') & (df['Heart rate'] != '?') & (df['P'] != '?') & (df['T'] != '?')]
#print(dfmean.median())
df['J'].replace({'?' : dfmean.median()['J']}, inplace=True)
df['QRST'].replace({'?' : dfmean.median()['QRST']}, inplace=True)
df['Heart rate'].replace({'?' : dfmean.median()['Heart rate']}, inplace=True)
df['P'].replace({'?' : dfmean.median()['P']}, inplace=True)
df['T'].replace({'?' : dfmean.median()['T']}, inplace=True)
df.head()
```

```python
#normalisasi
cols_to_scale = ['age', 'height', 'weight', 'QRS duration', 'P-R Interval', 'Q-T Interval', 'T Interval', 'P interval', 'QRS', 'T', 'P', 'QRST', 'J', 'Heart rate', 'y']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
df.head()
```

```python
#mengatur data train dan data test
X = df.drop('sex',axis='columns')
y = df['sex']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.18,random_state=5)
```

```python
print("ukuran data training x")
X_train.shape
```

```python
print("ukuran data test x")
X_test.shape
```

```python
#ALGORITMA ANN
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(15, input_shape=(15,), activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])


model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
            
#training
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
#evaluasi
model.evaluate(X_test, y_test)
```

```python
#prediksi
prediksi = model.predict(X_test)
prediksi[:10]
```

```python
y_prediksi = []
for element in prediksi:
    if element > 0.5:
        y_prediksi.append(1)
    else:
        y_prediksi.append(0)
y_prediksi[:10]
```

```python
y_test[:10]
```

```python
#performance matrix
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_prediksi)

plt.figure(figsize = (10,8))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
